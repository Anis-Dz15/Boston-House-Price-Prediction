Min-max Normalisation

```python
cols = ['crim','zn','tax','black']
for col in cols:
  # Find the minimum and maximum of that colum
  minimum = min(df[col])
  maximum = max(df[col])
  df[col] = (df[col] - minimum) / (maximum - minimum )
```

```python
#create displot
fig,ax = plt.subplots(ncols=7,nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col,value in df.items():
  sns.distplot(value,ax=ax[index])
  index +=1
plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
# Standardization
from sklearn import preprocessing
scalar = preprocessing.StandardScaler()

# Fit our data
scaled_cols = scalar.fit_transform(df[cols])
scaled_cols = pd.DataFrame(scaled_cols,columns=cols)
scaled_cols.head()
```

```python
# Affect the new data in the original dataframe
for col in cols:
  df[col] = scaled_cols[col]
```

```python
#create displot
fig,ax = plt.subplots(ncols=7,nrows=2, figsize=(20,10))
index = 0
ax = ax.flatten()

for col,value in df.items():
  sns.distplot(value,ax=ax[index])
  index +=1
plt.tight_layout(pad=0.5,w_pad=0.7,h_pad=5.0)
```

```python
X= df.drop(columns=['medv','rad'],axis=1)
y =df['medv']
```

Model training

```python
from sklearn.model_selection import cross_val_score,train_test_split
from sklearn.metrics import mean_absolute_error

def train(model,X,y):
  # Train the model
  X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42)
  model.fit(X_train,y_train)

  # Predict the traning set
  pred = model.predict(X_test)

  # Perform cross validation
  cv_score = cross_val_score(model,X,y,scoring='neg_mean_squared_error',cv=5)
  cv_score = np.abs(np.mean(cv_score))

  print('Model Report')
  print("MSE:",mean_absolute_error(y_test,y,pred))
  print('CV Score',cv_score)
```

```python
# Import necessary libraries
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
import pandas as pd

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
model = LinearRegression()
model.fit(X_scaled, y)
coef = pd.Series(model.coef_, index=X.columns).sort_values()
coef.plot(kind='bar', title='Model Coefficients')
```
